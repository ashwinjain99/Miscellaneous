```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
batting_df = pd.read_excel("Batting_Stats.xlsx")
batting_df
```

```python
batting_df.describe()
```

```python
batting_df.dropna(how="any")
```

```python
batting_df.dtypes
```

```python
batting_df["Age"] = batting_df["Age"].astype("int64")
```

```python
team_hr = batting_df.groupby("Tm")["HR"].sum()

top_hr_teams = team_hr.sort_values(ascending=False)
top_hr_teams.head()
```

```python
ATL_df = batting_df[batting_df["Tm"] == "ATL"].sort_values(by="HR", ascending=False).head(10)


plt.figure(figsize=(20,5))
plt.bar(ATL_df["Name"], ATL_df["HR"])
plt.xticks(rotation = 90)

plt.tight_layout()
plt.show()
```

```python
player_hr = batting_df.groupby("Name")["HR"].sum()
player_hits = batting_df.groupby("Name")["H"].sum()
player_runs = batting_df.groupby("Name")["R"].sum()
player_rbi = batting_df.groupby("Name")["RBI"].sum()
player_strikeouts = batting_df.groupby("Name")["SO"].sum()


top_player_hr = player_hr.sort_values(ascending=False)
top_player_hr.head()
```

```python
player_df = pd.DataFrame({"Total Home Runs": player_hr,
                          "Most Hits": player_hits,
                          "Most Runs": player_runs,
                          "Most RBIs": player_rbi})

player_df
```

```python
player_df[player_df["Total Home Runs"] >= 25]
```

```python
most_strikeouts = player_strikeouts.sort_values(ascending=False)
most_strikeouts.head()
```

```python
fig, ax = plt.subplots(2, figsize = (20, 6))

ax[0].scatter(batting_df["AB"], batting_df["HR"], marker = 'o', facecolor = "blue")

ax[1].scatter(batting_df["AB"], batting_df["H"], marker = 'o', facecolor = 'green')

plt.show()
```

```python
fig, ax = plt.subplots(2, figsize = (20, 6))

ax[0].scatter(batting_df["G"], batting_df["HR"], marker = "o", edgecolor = 'black')

ax[1].scatter(batting_df["G"], batting_df["H"], marker = "o", edgecolor = 'grey')

plt.show()
```
